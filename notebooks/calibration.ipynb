```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from pathlib import Path
from pprint import pprint

root_dir = Path(os.path.dirname(os.getcwd()))
print(root_dir)
```

```python
camera_id = "43"
year = "2024"
month = "08"
day = "29"

image_dir = Path(f"/mnt/gsdata/projects/ecosense/AngleCam2_0/data/raw_frames/G5Bullet_{camera_id}/{year}/{month}/{day}")
output_dir = Path(f"/mnt/gsdata/projects/ecosense/AngleCam2_0/data/corrected_frames/G5Bullet_{camera_id}/{year}/{month}/{day}")
images = list(image_dir.glob('*.jpg'))

pprint(images)
```

```python
def calibration(root_dir, img):
    mtx = np.load(root_dir / "data" / "calibration_matrix" / "mtx.npy")
    dist = np.load(root_dir / "data" / "calibration_matrix" / "dist.npy")
    calibrated_image = cv2.undistort(img, mtx, dist, None, mtx)
    return calibrated_image
```

```python
for fname in images[0:10]:
    img = cv2.imread(fname)
    undistorted = calibration(root_dir, img)
    filename = fname.stem + "_calibrated.jpg"
    cv2.imwrite(output_dir / filename , undistorted)
    print(f"Saved {output_dir / filename}")
```
